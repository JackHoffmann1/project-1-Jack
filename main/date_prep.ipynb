```python
# import libraries
import pandas as pd
```

```python
# Creating base dataframe from source
file = "Resources/card_transdata.csv"
df = pd.read_csv(file)
df.head()
```

```python
# Check column dtypes
df.info()
```

```python
# Check dataframe row and column length
df.shape
```

```python
# Check general statistics for first 4 columns
df.iloc[:,0:4].describe()
```

```python
# Change columns to a more user friendly name.
df.columns = ["Distance From Home","Distance From Last Transaction","Ratio to Median Purchase Price","Repeat Retailer","Used Chip","Used Pin Number","Online Order","Fraud"]
df.head()
```

```python
# Converted unnecessary float values to integers to represent binery 0 or 1 (True or False) statement.

df['Repeat Retailer'] = df['Repeat Retailer'].astype(int)
df['Used Chip'] = df['Used Chip'].astype(int)
df['Used Pin Number'] = df['Used Pin Number'].astype(int)
df['Online Order'] = df['Online Order'].astype(int)
df['Fraud'] = df['Fraud'].astype(int)
df.head()
```

```python
# Round float columns to two decimal places.

df = df.round(2)
df.head()
```

```python
# Save dataframe as csv file to Resources.

df.to_csv("Resources/cleaned_card_data.csv",index=False)
```
